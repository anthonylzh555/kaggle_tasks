```python
#drop
train_set.drop(['PM2.5'], axis=1, inplace=True)
#merge = pd.concat([new_train_set,new_test_set], axis=0)
#merge.reset_index(drop=True).head();
train_set.reset_index(drop=True).head();
test_set.reset_index(drop=True).head();

merge = pd.concat([train_set, test_set], axis=0)
#merge=merge.drop(['Date'],axis=1)
merge.reset_index(drop=True).head();
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
new_lat = le.fit_transform(merge.lat)
merge.lat = new_lat
new_lon = le.fit_transform(merge.lon)
merge.lon = new_lon
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
merge['lat'] = scaler.fit_transform(merge['lat'].values.reshape(-1,1))
merge['lon'] = scaler.fit_transform(merge['lon'].values.reshape(-1,1))
```

```python
train_set, test_set = np.split(merge,[7214], axis=0)
```

```python
train_set
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(train_set, class_train, test_size=0.2, random_state= 12, shuffle=False)
print(X_train.shape,X_valid.shape)
print(y_train.shape,y_valid.shape)
```

```python
from sklearn.linear_model import LinearRegression #選擇模型
from sklearn import metrics

model1 = LinearRegression()#建立模型
model1.fit(X_train,y_train)#訓練模型
print('train evaluation :')
pred = model1.predict(X_valid)#模型預測
print('R2 score:%.2f' %metrics.r2_score(y_valid, pred))
```

```python
from sklearn.linear_model import Lasso,Ridge

model4 = Lasso(alpha = 0.01)
model4.fit(X_train,y_train)
pred = model4.predict(X_valid)
print('train evaluation :')
print('R2 score:%.2f'%metrics.r2_score(y_valid, pred))
```

```python
model5 = Ridge(alpha = 0.01)
model5.fit(X_train,y_train)
pred = model5.predict(X_valid)
print('train evaluation :')
print('R2 score:%.2f'%metrics.r2_score(y_valid, pred))
#使用ridge驗證
```
